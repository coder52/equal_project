## Notwendige Importe

## Daten laden und vorbereiten

```python
data_path = 'Bewässerungszeiten-Datensatz/Bewässerungszeiten-Datensatz_small_31722.csv'
rides = pd.read_csv(data_path)
```

```python
dummy_fields = ['season', 'b12h_weathersit', 'b9h_weathersit', 'b6h_weathersit', 'b3h_weathersit', 
                'crnt_weathersit', 'a3h_weathersit', 'a6h_weathersit', 'a9h_weathersit', 'a12h_weathersit', 
                'mnth', 'hr']


for each in dummy_fields:
    dummies = pd.get_dummies(rides[each], prefix=each, drop_first=False)
    rides = pd.concat([rides, dummies], axis=1)

fields_to_drop = ['instant',  'season', 'b12h_weathersit', 'b9h_weathersit', 'b6h_weathersit', 'b3h_weathersit', 
                'crnt_weathersit', 'a3h_weathersit', 'a6h_weathersit', 'a9h_weathersit', 'a12h_weathersit', 
                'mnth', 'hr']

data = rides.drop(fields_to_drop, axis=1)
data.head()
```

```python
# shuffle the DataFrame rows
data = data.sample(frac = 1)
data.head()
```

## Daten in Trainings-, Validierungs- und Testsätze aufteilen

```python
# Save data for approximately the last 21 days 
test_data = data[-21*24:]

# Now remove the test data from the data set 
data = data[:-21*24]

# Separate the data into features and targets
target_fields = ['conclusion']
features, targets = data.drop(target_fields, axis=1), data[target_fields]
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields]
```

```python
features.head()
```

```python
# Hold out the last 60 days data or so of the remaining data as a validation set
train_features, train_targets = features[:-60*24], targets[:-60*24]
val_features, val_targets = features[-60*24:], targets[-60*24:]
```

```python
# Activation (sigmoid) function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_prime(x):
    return sigmoid(x) * (1-sigmoid(x))
def error_formula(y, output):
    return - (y*np.log(output) + (1 - y) * np.log(1-output))
def error_term_formula(x, y, output):
    return (y - output)*sigmoid_prime(x)
```

## Netzwerk-Trainer-Funktion

```python
# Training function
def train_nn(features, targets, learnrate, weights):

    n_records, n_features = features.shape
    last_loss = None

#     # Initialize weights
#     np.random.seed(42)
#     weights = np.random.normal( scale = 1 / n_features**.5, size=n_features )
    
    del_w = np.zeros(weights.shape)
    for x, y in zip(features, targets):
        
        output = sigmoid(np.dot(x, weights))
            
        # The error, the target minus the network output
        error = error_formula(y, output)

        # The error term
        error_term = error_term_formula(x, y, output)

        # The gradient descent step, the error times the gradient times the inputs
        del_w += error_term * x

    # Update the weights 
    weights += learnrate * del_w / n_records
        
    return weights
```

## Network Runner

```python
def run_nn(x, weights):
    output = sigmoid(np.dot(x, weights))
    return np.rint(output) # round to the nearest integer
```

## Trainiere das Netzwerk

```python
import sys
# Neural Network hyperparameters
epochs = 300
learnrate = 0.5

n_records, n_features = train_features.shape
# Use to same seed to make debugging easier
np.random.seed(42)
# Initialize weights
weights = np.random.normal( scale = 1 / n_features**.5, size=n_features )

for e in range(epochs):
    # train network and get new weights
    weights = train_nn(train_features.to_numpy(), train_targets.loc[:]['conclusion'].to_numpy(), learnrate, weights)
    # validate the network 
    output = run_nn(val_features.to_numpy(), weights)
    evaluation = output==val_targets.loc[:]['conclusion'].to_numpy()
    if e % 10 == 0:
        # mean square error on the training set    
        out = sigmoid(np.dot(train_features.to_numpy(), weights))
        train_loss = np.mean((out - train_targets.loc[:]['conclusion'].to_numpy()) ** 2)
        # mean square error on the validation set
        out = sigmoid(np.dot(val_features.to_numpy(), weights))
        val_loss = np.mean((out - val_targets.loc[:]['conclusion'].to_numpy()) ** 2)
        sys.stdout.write("\rEpoch: {:2d}".format(e) \
                     + " ... Training loss: " + str(train_loss)[:6] \
                     + " ... Validation loss: " + str(val_loss)[:6])
        sys.stdout.flush() 
```

## Netzwerk testen

```python
evaluation = []
for x, y in zip(test_features.to_numpy(), test_targets.loc[:]['conclusion'].to_numpy()):
    output = run_nn(x, weights)
    evaluation.append(output==y)
acc = np.sum(evaluation)/len(evaluation)
print("Test accuracy :", acc)
```
